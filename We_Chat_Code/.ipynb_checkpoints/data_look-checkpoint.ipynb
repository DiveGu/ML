```python
def statis_feature(df,dim='userid',start_day=1, before_day=7, agg='sum'):
    """
    统计用户/feed 过去n天各类行为的次数
    """
    print(dim)
    user_data = df[[dim, "date_"] + FEA_COLUMN_LIST]
    res_arr = []
    for start in range(start_day, END_DAY-before_day+1):
        # [1-7] [2-8] [3-9] ... [8-14]
        temp = user_data[(user_data["date_"] >= start) & (user_data["date_"] < (start + before_day))]
        temp = temp.drop(columns=['date_'])
        temp = temp.groupby([dim]).agg([agg]).reset_index()
        temp.columns = list(map('_'.join, temp.columns.values))
        temp["date_"] = start + before_day
        # print(temp) # temp是df，每一列是每个feature聚合后的结果
        res_arr.append(temp)
        
    dim_feature = pd.concat(res_arr)
    
    return dim_feature
```

```python
user_feature_sum=statis_feature(df_history,dim='userid',start_day=1, before_day=7, agg='sum')
user_feature_avg=statis_feature(df_history,dim='userid',start_day=1, before_day=7, agg='mean')
```

```python
# 对user的各种行为进行统计
def static_user_history(df,col_list,dim,day=14):
    df_need=df[df['date_']<day]
    df_static=pd.DataFrame()
    # 统计cnt
    temp = pd.DataFrame(df_need[[dim]+col_list].groupby(dim).agg('sum'))
    df_static[dim]=list(temp.index)
    for col in col_list:
        df_static['{}_{}_sum'.format(dim[0:-2],col)]=list(temp[col]) # -2是去掉id str
    # 统计mean
    temp = pd.DataFrame(df_need[[dim]+col_list].groupby(dim).agg('mean'))
    for col in col_list:
        df_static['{}_{}_mean'.format(dim[0:-2],col)]=list(temp[col])
        
    df_static['date_']=day
    return df_static
        
```

### 构造用户的近期行为

```python
df_history=pd.read_csv(USER_ACTION)
```

```python
# 对于dim1，取其对dim2的action近期行为List，长度为num
def process_user_history(df,day,actions,num,dim1,dim2):
    df_history=pd.DataFrame()
    uids=set(list(df[dim1]))
    uids=sorted(uids)
    sample_dict=dict()
    for a in actions:
        sample_dict['{}_{}_history'.format(dim1[:-2],a)]=[]
        
    for uid in uids:
        tmp=df[(df[dim1]==uid)&(df['date_']<day)]
        tmp=tmp.sort_values(by='date_',ascending=False)  
        for action in actions:
            try:
                cur_tmp=tmp[tmp[action]==1]
                if(len(cur_tmp)>=num):
                    ret=list(cur_tmp[dim2])[:num]
                    sample_dict['{}_{}_history'.format(dim1[:-2],action)].append(str(ret))
                else:
                    ret=list(cur_tmp[dim2])+[0]*(num-len(cur_tmp))
                    sample_dict['{}_{}_history'.format(dim1[:-2],action)].append(str(ret))
            except:
                    sample_dict['{}_{}_history'.format(dim1[:-2],action)].append(str([0]*num))
        
    df_history[dim1]=list(uids)
    for k,v in sample_dict.items():
        df_history[k]=v
    df_history['date_']=day
    return df_history
                        
        
```

```python
t0=time.time()
df_user_history=process_user_history(df_history,day=14,actions=['read_comment','like','click_avatar','forward'],num=10,dim1='userid',dim2='feedid')
t1=time.time()
print('process user history cost {:.2f} s'.format(t1-t0))
```

### 构造用户近期浏览行为 因为正样本太稀疏 

```python
# 对于dim1，取其对dim2的action近期行为List，长度为num
def process_user_history(df,num,dim1,dim2):
    uids=set(list(df[dim1]))
    uids=sorted(uids)
    sample_dim1=[]
    sample_his=[]
    sample_date=[]
    
    # 对于每一天滑窗统计所有uid之前的记录
    for day in range(1,16):
        print('process day={}'.format(day))
        cur_df=df[(df['date_']<day)&(df['date_']>=day-5)]
        # 对于每个user来选出其历史记录
        for uid in uids:
            try:
                # 先倒叙排列
#                 tmp=cur_df[cur_df[dim1]==uid].sort_index(ascending=False,inplace=False)
                tmp=cur_df[cur_df[dim1]==uid].sort_values(by=ACTION_LIST+['play'],ascending=[False]*4+[False])
                # 取最近num个
#                 if(len(tmp)>=num):
# #                     ret=list(tmp[dim2])[:num]
#                     ret=list(tmp[dim2])[::-1][:num] # 取倒叙
#                 else:
#                     ret=list(tmp[dim2])[::-1]+[0]*(num-len(tmp))
#                 ret=ret[::-1]

                # 正序排列
                if(len(tmp)>=num):
                    ret=list(tmp[dim2])[:num] #
                else:
                    ret=[0]*(num-len(tmp))+list(tmp[dim2])
                
                sample_dim1.append(uid)
                sample_his.append(str(ret))
                sample_date.append(day)
            except:
                ret=[0]*num
                sample_dim1.append(uid)
                sample_his.append(str(ret))
                sample_date.append(day)
    
    df_result=pd.DataFrame()
    df_result[dim1]=sample_dim1
    df_result['{}_watch_history'.format(dim1[:-2])]=sample_his
    df_result['date_']=sample_date
    
    return df_result
                        
```

```python
# print(df_feed_embed.loc[0,'feed_embedding'])
# 1 把每个feed的嵌入str转为np

def convert_2_np(df):
    emb=[]
    for tmp in df['feed_embedding']:
        emb.append([float(b) for b in tmp.split(' ')[:512]])
    emb=np.array(emb)
    return emb

emb=convert_2_np(df_feed_embed)
print(emb.shape)
# 2 进行聚类
```

```python
# 2 进行聚类
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

def get_feed_cluter(emb,n_clusters):
    kmeans = KMeans(init="k-means++", n_clusters=n_clusters, n_init=4,
                    random_state=0)
    estimator = make_pipeline(StandardScaler(), kmeans).fit(emb)
    cluter_label=estimator[-1].labels_
    return list(cluter_label)
```

```python
df_emb_cluter=pd.DataFrame()
df_emb_cluter['feedid']=df_feed_embed['feedid']
df_emb_cluter['feed_cluter']=get_feed_cluter(emb,100)
print(df_emb_cluter['feed_cluter'].value_counts())
```

```python
# 处理text列 截取max_size 保存成逗号隔开的str
def process_description(df,col,max_size=100):
    txt_list=[]
    for tmp in df[col]:
        try:
            tmp1=[int(w) for w in tmp.split(' ')]
            if(len(tmp1)>max_size):
                tmp1=tmp1[:max_size]
            else:
                tmp1=tmp1+[0]*(max_size-len(tmp1))
            tmp1=str(tmp1)
            txt_list.append(tmp1)
        except:
            txt_list.append([0]*max_size)
    
    return txt_list
```

```python
# 处理标签列
def process_tag(df,col,max_size):
    txt_list=[]
    txt_lens=[]
    for tmp in df[col]:
        try:
            # str是为了保存成 [id1,id2,id3]
            # +1是为了使得0作为填充
            if(col=='machine_tag_list'):
                tmp1=[int(w.split(' ')[0])+1 for w in tmp.split(';')]
            else:
                tmp1=[int(w)+1 for w in tmp.split(';')]
                
            if(len(tmp1)>max_size):
                tmp1=tmp1[:max_size]
            else:
                tmp1=tmp1+[0]*(max_size-len(tmp1))
            txt_list.append(str(tmp1))
            txt_lens.append(len(tmp1))
        except:
            txt_list.append([0]*max_size)
            pass
    
    
    # 统计标签长度的分布
    df_lens=pd.DataFrame()
    df_lens['txt_lens']=txt_lens
    rank=list(np.arange(0,1.1,0.1))
    ret=df_lens.quantile(rank)
    df_cnt=pd.DataFrame()
    df_cnt['lens_cnt']=df_lens['txt_lens'].value_counts()
    df_cnt['lens_cnt_ratio']=df_cnt['lens_cnt']/len(df_lens)*100
    print(df_cnt)
    
    return txt_list
    
```

```python
# 标签处理
df_feed_feature['manual_tag']=process_tag(df_feed_feature,'manual_tag_list',4)
df_feed_feature['machine_tag']=process_tag(df_feed_feature,'machine_tag_list',6)
```
