```python
import numpy as np
import pandas as pd
import time
from sklearn import preprocessing
```

```python
# 存储数据的根目录
ROOT_PATH = "/testcbd017_gujinfang/GJFCode/WeChat_2021/Code/data"
SUB_PATH=ROOT_PATH+'/submit'
```

```python
subset={
#     SUB_PATH+'/sub_wdl_des_tag_keyword_pair_8.csv':5,
        SUB_PATH+'/b_tree_1.csv':[4,4.4,7,7],
        SUB_PATH+'/b_tree_2.csv':[4,4.4,7,7],
        SUB_PATH+'/b_deepfm1_3.csv':[3.5,0,0,0],
        SUB_PATH+'/b_deepfm2_3.csv':[3.5,0,0,0],
        SUB_PATH+'/b_sub_nn_1_5.csv':[1,1,1.1,1.1],
        SUB_PATH+'/b_sub_nn_2_5.csv':[1,1,1.1,1.1],
        SUB_PATH+'/b_sub_nn_3_5.csv':[1,1,1.1,1.1],
        SUB_PATH+'/b_sub_nn_4_5.csv':[1,1,1.1,1.1],
        SUB_PATH+'/b_sub_nn_5_5.csv':[1,1,1.1,1.1],
        SUB_PATH+'/b_sub_nn_6_5.csv':[1,1,1.1,1.1],
        SUB_PATH+'/b_sub_nn_7_5.csv':[1,1,1.1,1.1],
        SUB_PATH+'/b_sub_nn_8_5.csv':[1,1,1.1,1.1],
}
```

```python
submit_final=pd.DataFrame()
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0., 1.))
cols=["read_comment", "like", "click_avatar",  "forward"] # 需要归一化的列
base_cols=['userid','feedid']
```

```python
# # 归一化提交结果权重
# weights=list(subset.values())
# weights=[w/sum(weights) for w in weights]
# subset=dict(zip(subset.keys(),weights))
# # print(subset)
```

```python
values_cols=[]
# 计算融合结果
for file_path,weight in subset.items():
    print(weight)
    cur_submit=pd.read_csv(file_path)
    submit_final[base_cols]=cur_submit[base_cols]
    cur_values=min_max_scaler.fit_transform(cur_submit[cols].values)
#     print(cur_values[:5,:])
    if(values_cols==[]):
        values_cols=np.array([[0.]*len(cols)]*len(cur_submit))
    print(cur_values)
    values_cols+=cur_values*weight
# print(values_cols[:5,:])
```

```python
print(values_cols)
```

```python
submit_final[cols]=min_max_scaler.fit_transform(values_cols)
```

```python
save_path='{}/b_sub_nn_tree_{}model_{}.csv'.format(SUB_PATH,len(subset),'1')
submit_final.to_csv(save_path,index=False)
```
